```python
#Converting songs duration from milli seconds to seconds
df['Duration']=df['duration_ms'].apply(lambda x:round(x/1000))
df.drop("duration_ms",axis=1,inplace=True)
```

```python
df['Duration'].head()
```

```python
#Correlation heatmap between Variables by taking away unneccessary values
corr_df = df.drop(["key","mode","genre","artist_name","track_name","track_id","time_signature"],axis=1).corr(method="pearson")
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="coolwarm")
heatmap.set_title("Correlation heatmap between Varirables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
numeric_df = df.select_dtypes(['int64','float64'])
numeric_df
```

```python
#Correlation between variables

corr_df=numeric_df.corr(method="pearson")
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="coolwarm")
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels())
```

```python
#Converting time_signature from fraction form to decimal form

def convert_time_Signature(ts):
  numerator,denominator=ts.split("/")
  return int(numerator)/int(denominator)

df['time_signature_decimal'] = df['time_signature'].apply(convert_time_Signature)
df
```

```python
df.rename(columns={"time_signature_decimal":"time_signature"},inplace=True)
```

```python
df.head(2)
```

```python
#Changing the mode of Minor and Major as 0's and 1's for better data analysis
def conv_mode(value):
  if value=="Minor":
    return 0
  else:
    return 1

df['mode']=df['mode'].apply(conv_mode)
```

```python
#to know 0.4% of the total rows in the dataset
sample_df = df.sample(int(0.004*len(df)))
sample_df.shape
```

```python
#Regresssion plot to know the correlation between loudness and energy

sns.regplot(x="loudness",y="energy",data=sample_df,color="c")
plt.title("Loudness vs Energy")
plt.show()
```

```python
#Regresssion plot to know the correlation between Popularity and acousticness

sns.regplot(x="popularity",y="acousticness",data=sample_df)
plt.title("Popularity vs acousticness")
plt.show()
```
